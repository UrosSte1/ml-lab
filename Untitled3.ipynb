```python
from sklearn.model_selection import train_test_split
import numpy as np
#df.rename({"X3*": "X3", "Y*" : "Y"})
data = df[['partial paresis', 'class']]
data = data.rename(columns={"partial paresis":"X", 'class': 'Y'})
train, test = train_test_split(data, test_size=0.2, random_state=42)

# racunamo beta
x_mean = np.mean(train["X"])
y_mean = np.mean(train["Y"])
brojilac = 0
imenilac = 0
for index, row in train.iterrows():
  x = row[0]
  y = row[1]
  brojilac += (x - x_mean)*(y-y_mean)
  imenilac += (x - x_mean) ** 2

beta = brojilac/imenilac


alpha = y_mean - beta * x_mean
```

```python

import math
def rmse(y_true, y_pred):
  sum = 0
  for yt, yp in zip(y_true, y_pred):
    sum += (yt - yp) ** 2
  return math.sqrt(sum / len(y_true))

```

```python
model = lambda x, alpha, beta: alpha + beta * x
y_pred = model(test['X'], alpha, beta)
y_true = test['Y']

print(rmse(y_true, y_pred))
```
